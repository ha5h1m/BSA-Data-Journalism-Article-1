***

Create a table for home fixture data that just has Team, Home Scored, and Home Conceded columns

Use groupby function to find each team's averages for these columns

```python
homeStats = df[['HomeTeam','FTHG','FTAG']].rename(
    columns={'HomeTeam':'Team','FTHG':'HomeScored','FTAG':'HomeConceded'})
homeStats
```

```python
homeAverage = homeStats.groupby(['Team'],as_index=False)[['HomeScored','HomeConceded']].mean()
homeAverage
```

Create a Goal Difference table for homefixtures by minusing 'HomeScored' column from 'HomeConceded'

```python
GoalDifference = homeAverage['HomeScored']-homeAverage['HomeConceded']
homeAverage['GDhome']=GoalDifference
homeAverage
```

***

Repeat the same process with away fixture data 

```python
newhomeAverage = homeAverage.drop(['HomeScored','HomeConceded'],axis=1)
newhomeAverage
```

```python
awayStats=df[['AwayTeam','FTHG','FTAG']].rename(columns={'AwayTeam':'Team','FTHG':'AwayConceded','FTAG':'AwayGoals'})
awayStats
```

```python
awayAverage=awayStats.groupby(['Team'],as_index=False)[['AwayGoals','AwayConceded']].mean()
awayAverage
```

```python
AwayGD = awayAverage['AwayGoals']-awayAverage['AwayConceded']
awayAverage['GDaway']=AwayGD
awayAverage
```

```python
newawayAverage=awayAverage.drop(['AwayGoals','AwayConceded'],axis=1)
newawayAverage
```

***

Merge the Home and Away goal difference tables together to create one table

```python
HomeAwayGD = pd.merge(newhomeAverage,newawayAverage,on='Team')
HomeAwayGD.set_index('Team',inplace=True)
HomeAwayGD
```

***

Make match predictor function that simulates a match and calculate how many points the Home and Away team gets

```python
def MatchPredictor(home, away):
    if home in HomeAwayGD.index and away in HomeAwayGD.index:
        pH, pA  = 0, 0 
        H = HomeAwayGD.at[home,'GDhome'] 
        A = HomeAwayGD.at[away,'GDaway']
        if H > A+0.6:
            pH += 3
            pA += 0
        elif A > H+0.6:
            pA += 3   
            pH += 0 
        elif H - A < 0.6 or A - H < 0.6:
            pH += 1
            pA += 1
        return (pH, pA)
    else:
        return (0,0)
```

Test that the function works for some random fixture

```python
MatchPredictor('Arsenal', 'Bournemouth')
```

***

Apply function so that it iterates through our remaining matches spreadsheet and updates our current points table in the process

```python
for index, row in RemainingMatches.iterrows():
    home, away = row['HomeTeam'], row['AwayTeam']
    
    sH, sA = MatchPredictor(home, away)
    plTable.loc[plTable.Team == home, 'Points'] += sH
    plTable.loc[plTable.Team == away, 'Points'] += sA
```

```python
Predicted_Table = plTable.sort_values('Points', ascending=False)
plTable.index = np.arange(1, len(plTable)+1) 
```
